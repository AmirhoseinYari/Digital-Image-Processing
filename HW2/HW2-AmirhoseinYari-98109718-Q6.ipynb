```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_rgb = cv2.imread('q7_1.png')
img = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
pattern = cv2.imread('q7_2.png',0)
w, h = pattern.shape[::-1]
```

## Cross Correlation

```python
img_rgb = cv2.imread('q7_1.png')
img = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)

res1 = cv2.matchTemplate(img, pattern, cv2.TM_CCORR)
threshold1 = 0.99*np.max(res1)
loc1 = np.where( res1 >= threshold1)

for pt in zip(*loc1[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 5)
    
plt.figure(figsize=(20,10))

plt.subplot(121),plt.imshow(res1,cmap = 'gray')
plt.title('Matching Result'), plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB),cmap = 'gray')
plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    
cv2.imwrite('q7_res01.jpg',img_rgb)
```

## NCC(Normalized Cross Corelation)

```python
img_rgb = cv2.imread('q7_1.png')
img = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)

res2 = cv2.matchTemplate(img, pattern, cv2.TM_CCORR_NORMED)
threshold2 = 0.99*np.max(res2)
loc2 = np.where( res2 >= threshold2)

for pt in zip(*loc2[::-1]):
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 5)
    
plt.figure(figsize=(20,10))

plt.subplot(121),plt.imshow(res2,cmap = 'gray')
plt.title('Matching Result'), plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB),cmap = 'gray')
plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    
cv2.imwrite('q7_res02.jpg',img_rgb)
```
