```python
import imageio as iio
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img = cv2.imread('q4.png',0)
#plt.imshow(img,cmap='gray');
```

```python
res1 = cv2.equalizeHist(img)

# create a CLAHE object
clahe = cv2.createCLAHE(clipLimit=20, tileGridSize=(3,3))
res2 = clahe.apply(img)

clahe = cv2.createCLAHE(clipLimit=1, tileGridSize=(3,3))
res3 = clahe.apply(img)

plt.figure(figsize = (16, 12))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray');
plt.title('image :');

plt.subplot(2,2,2)
plt.imshow(res1,cmap='gray');
plt.title('HE :');

plt.subplot(2,2,3)
plt.imshow(res2,cmap='gray');
plt.title('AHE :');

plt.subplot(2,2,4)
plt.imshow(res3,cmap='gray');
plt.title('CLAHE :');


iio.imwrite('q4_1.png',res1)
iio.imwrite('q4_2.png',res2)
iio.imwrite('q4_3.png',res3)
```
