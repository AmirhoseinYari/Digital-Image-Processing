```python
import cv2
import cmath
import numpy as np
import imageio as iio
from matplotlib import pyplot as plt
```

```python
hand = cv2.imread('hand_xray.jpg',0)
#brain = cv2.imread('brain_xray.jpg',0)
brain = iio.imread('brain_xray.jpg')

#print(hand.shape, brain.shape)
```

```python
handF = np.fft.fftshift(np.fft.fft2(hand))
brainF = np.fft.fftshift(np.fft.fft2(brain))

#img1F = np.multiply(np.abs(brainF),np.exp(1j*np.angle(handF)))
img1F = np.abs(brainF)*np.exp(1j*np.angle(handF))
img2F = np.multiply(np.abs(handF),np.exp(1j*np.angle(brainF)))

img1 = np.real(np.fft.ifft2(np.fft.ifftshift(img1F)))
img2 = np.real(np.fft.ifft2(np.fft.ifftshift(img2F))) ########
```

```python
plt.figure(figsize = (18,10))

plt.subplot(1,4,1)
plt.imshow(hand,cmap='gray')
plt.title('hand_xray :')
plt.subplot(1,4,2)
plt.imshow(img1,cmap='gray')
plt.title('brain amplitude + hand phase :')

plt.subplot(1,4,3)
plt.imshow(brain,cmap='gray')
plt.title('brain_xray :')
plt.subplot(1,4,4)
plt.imshow(img2,cmap='gray')
plt.title('hand amplitude + brain phase :');
```
